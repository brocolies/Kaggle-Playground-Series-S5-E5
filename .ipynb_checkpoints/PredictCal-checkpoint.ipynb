```python
def add_features(df): # test_df에 일관되게 적용하기 위해 함수 생성 
    df = df.copy() # 원본 보호 
    
    train_df['IsAge48'] = (train_df['Age'] == 48).astype(int)
    train_df['IsAge76_78'] = (train_df['Age'].isin([76, 77, 78])).astype(int)
    train_df['IsTaller210'] = (train_df['Height'] > 210).astype(int)
    train_df['BMI'] = (train_df['Weight'] / (train_df['Height'] / 100) ** 2).round(0)
    train_df['IsBMIHigher40'] = (train_df['BMI'] >= 40).astype(int)
    train_df['IsWeightHigher112'] = ((train_df['Weight'] >= 112) & (train_df['Weight'] <= 126)).astype(int)
    train_df['Heart_Rate'] = train_df['Heart_Rate'].clip(lower=70) # 70보다 작은 애들은 70으로 고정
    train_df['Intensity'] = (train_df['Heart_Rate'] - 70) + ((train_df['Body_Temp'] - 34) * 6).clip(lower=23)
    train_df['EnergyProxy'] = train_df['Duration'] * (train_df['Intensity'] - 23)
    train_df['Intensity'] = train_df['Intensity'].clip(lower=23) # 23보다 작은 애들은 23으로 고정

    return df 
```
